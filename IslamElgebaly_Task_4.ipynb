```python
import pandas as pd
import numpy as np
import user_agents
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("drive/MyDrive/Datasets/visitor-interests.csv")
df = df.head(50)
df.sample(10)
```

```python
def extract_info(s):

  ua = user_agents.parse(s["UserAgent"])
  s["Browser"] = ua.browser.family
  s["OS"] = ua.os.family
  s["Device"] = ua.device

  return s
```

```python
df = df.apply(extract_info, axis = 1)
```

```python
df[["Browser", "OS", 'Device']]
```

**2) Using visitor-interests.csv dataset, extract latitude, longitude from 'IP' column.**

```python
def extract_geo(s):
  ip = ip2geo.get(s["IP"])
  s["Latitude"] = ip.latitude
  s["Longitude"] = ip.longitude
  return s
```

```python
df = df.apply(extract_geo, axis = 1)
```
